```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset("mpg")
mpg
#full table of data (398 rows)
```

```python
mpg = mpg.dropna(subset=['horsepower', 'mpg'])
mpg
#cleaning missing data rows new updated table (392 rows)
```

```python
# Create a visual showing the relationship between horsepower and fuel economy.
plt.figure(figsize=(12, 6))
sns.lineplot(x='horsepower', y='mpg', data=mpg, hue='origin', style='origin', markers=True)
plt.xlabel('Horsepower')
plt.ylabel('Fuel Economy (mpg)')
plt.title('Relationship between Horsepower and Fuel Economy')
plt.show()

# This graph is showing how the horsepower can effect vehicles Fuel economy.
#I noticed USA will have stronger horsepower over 225 and other country's did not go past 130
# I see USA when they have stronger horsepower it decresses alot from there first peak at 60
# I see Europe started great but just like Usa started to decrease over time
# Japan started close to Usa but at there highist horsepower they almost made it back to there best at 60
```

```python
# Create a table summarizing fuel economy by country of origin
import pandas as pd
```

```python
mpg
```

```python
sum_of_the_table = mpg.groupby('origin')['mpg'].mean().reset_index()
```

```python
sum_of_the_table.columns = ["Country's", 'Average Fuel (MPG)']
print(sum_of_the_table)
```

```python
#Create a visual showing how fuel economy has changed over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='model_year', y='mpg', data=mpg, marker='o')
plt.xlabel('Model Year')
plt.ylabel('Fuel Economy (mpg)')
plt.title('Change in Fuel Economy Over Time')
plt.grid(True)
plt.show()

# I notice on this graph the newer the car more MPG is needed.
# The reason could be because more travel, bigger cars to fit bigger families, even stronger horsepower most likely.
```

```python
fuel_economy = mpg.groupby('origin')['mpg'].agg(['mean', 'median', 'std', 'min', 'max']).reset_index()
fuel_economy.columns = ['Country of Origin', 'Mean MPG', 'Median MPG', 'Standard Deviation', 'Min MPG', 'Max MPG']
print(fuel_economy)
```
